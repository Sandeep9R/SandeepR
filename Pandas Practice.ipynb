# typecasting columns

```python
def convertInt(x):
    if x == "Null":
        return 0
    else:
        return int(x)
```

```python
def convertFloat(x):
    if x == "Null":
        return 0.0
    else:
        return float(x)
```

```python
empdf.info()
```

```python
empdf.department_id=empdf.department_id.apply(convertInt)
```

```python
empdf.manager_id=empdf.manager_id.apply(convertInt)
```

```python
empdf.commission_pct=empdf.commission_pct.apply(convertFloat)
```
