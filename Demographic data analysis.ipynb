```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
DF = pd.read_csv("DemographicData.csv")
DF.columns = (['CountryName', 'CountryCode', 'BirthRate', 'InternetAccess',
       'IncomeGroup'])
```

```python
DF.head()
```

```python
sns.distplot(DF['InternetAccess'],color='red',hist_kws={'edgecolor':'black'})
plt.show()
```

```python
sns.scatterplot(DF['InternetAccess'],DF['BirthRate'])
```

```python
sns.distplot(DF[DF.IncomeGroup=="High income"].InternetAccess, 
             hist_kws={"edgecolor":"black"},
            kde_kws={"color":"Red","shade":True})
plt.show()
```

```python
sns.distplot(DF[DF.IncomeGroup=="Low income"].InternetAccess, 
             hist_kws={"edgecolor":"black"},
            kde_kws={"color":"Green","shade":True})
plt.show()
```

```python
sns.distplot(DF[DF.IncomeGroup=="High income"].InternetAccess, 
             hist_kws={"edgecolor":"black"},
            kde_kws={"color":"Red","shade":True})

sns.distplot(DF[DF.IncomeGroup=="Low income"].InternetAccess, 
             hist_kws={"edgecolor":"black"},
            kde_kws={"color":"Green","shade":True})
plt.show()
```
